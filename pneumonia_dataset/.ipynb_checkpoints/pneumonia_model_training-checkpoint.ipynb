```python
import os
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve
```

```python
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from tensorflow.keras.models import Sequential, load_model
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
dataset_path = "C:/Users/barra/Desktop/project/pneumonia_dataset"
```

```python
img_width, img_height = 128,128
```

```python
batch_size = 32
```

```python
epochs = 20
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1.0/255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    validation_split = 0.2
)
```

```python
train_generator = train_datagen.flow_from_directory(
    os.path.join(dataset_path, 'train'),
    target_size = (img_width, img_height),
    batch_size=batch_size,
    class_mode = 'binary',
    subset = 'training'
)
validation_generator = train_datagen.flow_from_directory(
    os.path.join(dataset_path, 'val'),
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)
```

define your model

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
checkpoint = ModelCheckpoint('best_weights.weights.h5', monitor='val_loss', save_best_only=True, mode='min', save_weights_only=True, verbose=1)

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_steps=validation_generator.samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    callbacks=[checkpoint]
)
model.save('pneumonia_model.h5')
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()
```

```python
# Additional visualizations
# 1. Learning rate
plt.figure(figsize=(12, 4))
plt.plot(history.history.get('lr', []))  # Use .get() to avoid KeyError if 'lr' is not present
plt.title('Learning Rate over Epochs')
plt.ylabel('Learning Rate')
plt.xlabel('Epoch')
plt.show()
```
